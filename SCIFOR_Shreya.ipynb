```python
import seaborn as sns
import matplotlib.pyplot as plt

iris = sns.load_dataset('iris')

# Bar Chart
plt.figure(figsize=(10, 6))
sns.countplot(x='species', data=iris, palette='viridis')
plt.title('Bar Chart - Iris Species Counts')
plt.xlabel('Species')
plt.ylabel('Count')
plt.show()

# Pie Chart
species_counts = iris['species'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(species_counts, labels=species_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Pie Chart - Distribution of Iris Species')
plt.show()

# Heatmap (using correlation matrix)
correlation_matrix = iris.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap - Correlation Matrix of Iris Dataset')
plt.show()

# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=iris, x='sepal_length', kde=True, color='skyblue')
plt.title('Histogram with KDE - Sepal Length Distribution')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.show()

# Boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(x='species', y='petal_length', data=iris, palette='Set3')
plt.title('Boxplot - Petal Length across Iris Species')
plt.xlabel('Species')
plt.ylabel('Petal Length')
plt.show()

# Scatterplot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris, palette='Set2')
plt.title('Scatterplot - Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.show()

```
